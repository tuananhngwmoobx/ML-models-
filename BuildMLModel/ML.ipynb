# House price prediction project
1. Problem definition:
   * Goal: predict the sales price for each house. For each Id in the test set, you must predict the value of the SalePrice variable. 

```python
import pandas as pd
import numpy as np
```

```python
data_df = pd.read_csv("train.csv")
```

```python
data_df.head()
```

```python
data_df = pd.read_csv("train.csv", index_col = "Id")
```

```python
data_df
```

```python
data_df.columns
```

### 2. Feature selection
* Choose important features to train ML model (Feature Engineering)

```python
features = ["LotArea", "YearBuilt", "1stFlrSF", "2ndFlrSF", "FullBath", "BedroomAbvGr", "TotRmsAbvGrd"]
```

### 3. Splitting the dataset
* x: data[feature]
* y: target variable SalePrice

```python
x = data_df[features]
```

```python
x
```

```python
y = data_df["SalePrice"]
```

```python
y
```

### Splitting dataset into training set and test set
## To avoid the overfitting
# x, y -> x_train, y_train, x_test, y_test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2, random_state=0)
#random_state: do not change the state anytime run the func
```

```python
x_train.shape
```

```python
x_test.shape
```

### 4.Training the ML model

#### 1. Using decision tree

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(random_state=1)
```

```python
# Fit training data into model
dt_model.fit(x_train, y_train)
```

```python
y_predicts = dt_model.predict(x_test.head())
```

```python
y_predicts
```

```python
pd.DataFrame({'y': y_test.head(), 'y_predicts': y_predicts})
```

#### The diff is still quite mass

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(x_train, y_train)
```

```python
y_rf_pred = rf_model.predict(x_test)
```

```python
y_rf_pred
```

```python
pd.DataFrame({'y': y_test, 'y_predicts': y_rf_pred})
```
