## Pyplot API

```python
x = np.linspace(0, 10, 1000)
```

```python
plt.title("Sin(x) and Cos(x)")
plt.plot(x, np.sin(x), color = "blue", linestyle = "dashed", label  ="sin(x)") 
plt.plot(x, np.cos(x), color = "red", linestyle = "dotted", label = "cos(x)") 
plt.xlabel("Biến x")
plt.ylabel("F(x)")
# plt.xlim([0, 4])
# plt.ylim([-0.5, 1.5])
plt.axis([0, 4, -0.5, 1])
plt.axis("tight")   #make the graph closer 
#plt.axis("equal")
plt.legend() #call labels

#plt.show() trong jupyter ko cần dùng hàm nãy, có thể dùng hàm này khi chạy ctr trong terminal
```

### Object-Oriented API


#### 1. Prepare data

```python
x = [1,2,3,4]
y = [11,22,33,44]
```

#### 2. Setup plot and Plot the data
#### Trong OO API, một figure sẽ có nhiều subplots đc specify bằng nrows và ncols
#### Trong đó có Figure and Axes objects 

```python
#specify số subplot 
fig, ax = plt.subplots( figsize = (5,5))  #figure size = Width and Height of the plot 
#fig là hình tổng còn ax là những hình nhỏ 
ax.plot(x, y);
ax.set(title = "A simple plot", xlabel = "x-axis", ylabel = "y-axis")
```
