## Pyplot API

```python
x = np.linspace(0, 10, 1000)
```

```python
plt.title("Sin(x) and Cos(x)")
plt.plot(x, np.sin(x), color = "blue", linestyle = "dashed", label = "sin(x)") 
plt.plot(x, np.cos(x), color = "red", linestyle = "dotted", label = "cos(x)") 
plt.xlabel("Bin x")
plt.ylabel("F(x)")
# plt.xlim([0, 4])
# plt.ylim([-0.5, 1.5])
plt.axis([0, 4, -0.5, 1])
plt.axis("tight")   #make the graph closer 
#plt.axis("equal")
plt.legend() #call labels

#plt.show() trong jupyter ko cn dng hm ny, c th dng hm ny khi chy ctr trong terminal
```

### Scatter

```python
#Pyplot API
plt.scatter(x, np.exp(x)); #y = e^x
```

```python
# OO API 
# Prepare for New Data
rng = np.random.RandomState(0)
x = rng.randn(100)   #From standard normal distribution
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000*rng.rand(100) 

fig, ax = plt.subplots()
img1 = ax.scatter(x, y, s = sizes, c = colors, cmap = 'viridis', alpha = 0.3) #cp x, y c kch thc khc nhau
#alpha: transparent 
fig.colorbar(img1)
```

### Bar chart
* Vertical chart
* Horizontal chart 

```python
#Prepare data from a Python dict
#Vertical chart
soft_drink_prices = {"Coke" : 10, "Pepsi" : 12, "Sprite" : 8}
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values())
ax.set(title = "WinMart's soft drinks prices", ylabel = "Price ($)")
```

```python
#Horizontal chart
fig, ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()))     #barh ch dng c vi list
```

#### Histogram

```python
#prepare data 
np.random.seed(42)  
student_height = np.random.normal(170, 10, 250)    #tb: 170, deviation: 10, 250 students
```

```python
student_height[:10]
```

```python
#Pyplot API
plt.hist(student_height, bins = 30) #bins: chia nh
```

### Subplots

```python
#Option 1: Must use OO API
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5)) #Width = 10, height = 5 

#ax1, ax2 on the same row, ax3,ax4 on the same row
#Plot data to each axis
ax1.plot(x, x/2)
ax1.plot(x, x*2)
ax2.scatter(np.random.random(10), np.random.random(10)) 
ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values())
ax4.hist(student_height, bins = 30)
```

```python
#Option 2: using OO API
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5))
ax[0, 0].plot(x, x/2)
ax[0, 0].plot(x, x*2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10)) 
ax[1, 0].bar(soft_drink_prices.keys(), soft_drink_prices.values())
ax[1, 1].hist(student_height, bins = 30)
```

### Data visualization in California cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('california_cities.csv')
cities.head()
cities ["area_total_km2"].sort_values(ascending = False)
#C th chia thnh tng khong 50 100 200 ....
#latitude: v : ngang
#longtitude: kinh : dc
#giao ca kinh v v l a im ca chng ta ang ng
```

```python
#Extract latitude and longtitude
lat, lon = cities['latd'], cities['longd']
lat[:10]
```

```python
lon[:10]
```

```python
#plot using pyplot API
population, area = cities['population_total'], cities['area_total_km2']
# plt.style.use('seaborn')
plt.figure(figsize = (8,6))
plt.scatter(lon, lat, c = np.log10(population), cmap = 'viridis', s = area, linewidths = 0, alpha = 0.5) #need log to modify the scale of population
plt.axis('equal')
plt.xlabel('longtitude')
plt.ylabel('ladtitude')
plt.colorbar(label = 'log$_{10}${population)')
plt.clim(3,7) #color bar: 3 - 7
plt.title('California cities: population and area distribution')

#Create a legend for cities size
area_range = [50, 100, 300, 500]
for area in area_range:
    plt.scatter([],[], s = area, label = str(area) + 'km$^2$')
plt.legend(scatterpoints = 1, labelspacing = 1, title = 'City Area')    
```
